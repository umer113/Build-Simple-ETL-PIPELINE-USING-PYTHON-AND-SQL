```python
import pandas as pd
from sqlalchemy import create_engine

# Step 1: Read the Excel file
df = pd.read_excel('D:\\data engineering\\Project1\\weather_forecast_data.xlsx')

# Step 2: Create a connection to the MySQL database
username = 'umer'
password = '3189'
host = 'localhost'
database = 'WEATHER_DB'
connection_string = f'mysql+pymysql://{username}:{password}@{host}/{database}'
engine = create_engine(connection_string)

# Step 3: Store the data in MySQL
table_name = 'Forecasts'
df.to_sql(table_name, con=engine, if_exists='replace', index=False)

# Step 4: Verify data in MySQL
query = f'SELECT * FROM {table_name}'
result_df = pd.read_sql(query, con=engine)
print(result_df.head())
```
